```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Melbourne_housing_FULL.csv')
data
```

```python
data = data.dropna()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Suburb'] = le.fit_transform(data['Suburb'])
data['Address'] = le.fit_transform(data['Address'])
data['Method'] = le.fit_transform(data['Method'])
data['SellerG'] = le.fit_transform(data['SellerG'])
data['CouncilArea'] = le.fit_transform(data['CouncilArea'])
data['Regionname'] = le.fit_transform(data['Regionname'])
data['Type'] = le.fit_transform(data['Type'])
```

```python
data
```

```python
x = data.drop('Price',axis=1)
x = np.array(x.drop('Date',axis=1))

y = np.array(data['Price'].astype(int))
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model = LinearRegression()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1,random_state=42)
model.fit(xtrain,ytrain)
```

```python
import sklearn
import sklearn.metrics
prediction = model.predict(xtest)
accurcy = model.score(xtest,ytest)
print(accurcy)
```

```python
from sklearn.ensemble import RandomForestRegressor

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1,random_state=42)
model = RandomForestRegressor()
model.fit(xtrain,ytrain)
```

```python
accurcy = model.score(xtest,ytest)
```

```python
print(accurcy)
```

```python
import pickle
with open('model.pickle','wb') as file:
    pickle.dump(model,file)
```

```python
from sklearn import svm

model = svm.SVR(kernel='linear')

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1,random_state=42)
model.fit(xtrain,ytrain)
```

```python
accuracy = model.score(xtest,ytest)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1,random_state=42)
model.fit(xtrain,ytrain)
```

```python
accuracy = model.score(xtest,ytest)
print(accuracy)
```

```python
with open('model.pickle','rb') as file:
    model = pickle.load(file)
print(model.score(xtest,ytest))
```
